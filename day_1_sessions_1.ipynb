## Sentiments analysis using transformer models


```python
from transformers import pipeline

classifier =pipeline("sentiment-analysis")
classifier(
    [
        "i've been waiting for a Huggingface course my whole life.",
        "I hate this so much"
    ]
)
```

## Text generation model(like chatgpt)

```python
generator=pipeline("text-generation")
generator("In this course, we will teach everything")
```

```python
qa=pipeline("question-answering")
qa(
    question="Where do I work?",
  context="My name is Tilak and i currently study at NCE"
)
```

## Language translator

```python
translator=pipeline("translation_en_to_de")
translator("I love programming.")
```

## NER models

```python
ner=pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")
ner("My name is Tilak and i currently study at NCE and i live in kathmandu")
```
